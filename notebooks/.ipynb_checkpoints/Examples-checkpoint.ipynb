```python
import pretty
import numpy as np
import matplotlib.pyplot as plt 

x1 = np.linspace(0,3,1000)
y1 = 0.2*(np.exp(x1) - x1**x1)

x2 = np.linspace(-2,2,1000)
y2 = np.sin(np.tan((x2)))
```

First a showcase of each of the included styles

```python
with pretty.LHCb():
    fig = plt.figure()    
#    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(1, 1, 1)
    
    ax.plot(x2,y2, label = 'a')
    ax.plot(x2, [z + .1 for z in y2], label = 'b')
    ax.plot(x2, [z + .2 for z in y2], label = 'c')
    ax.plot(x2, [z + .3 for z in y2], label = 'd')        
    ax.set_xlabel(r'$x$', ha='right', x=1)
    ax.set_ylabel(r'$tan(sin(...tan(sin(\gamma x))...))$', ha='right', y=1)
    ax.legend(loc='best')
    ax.minorticks_on()
    plt.show()
```

```python
with pretty.Classic():
    fig = plt.figure()    
#    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(1, 1, 1)
    
    ax.plot(x2,y2, label = r'a')
    ax.plot(x2, [z + .1 for z in y2], label = r'b')
    ax.plot(x2, [z + .2 for z in y2], label = r'c')
    ax.plot(x2, [z + .3 for z in y2], label = r'd')        
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$sin(\gamma x)$')
    ax.legend(loc='best')
    plt.show()
```

```python
with pretty.Monochrome():
    fig = plt.figure()    
#    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(1, 1, 1)
    
    ax.plot(x1,y1, label = 'a')
    ax.plot(x1, [z + .1 for z in y1], label = 'b')
    ax.plot(x1, [z + .2 for z in y1], label = 'c')    
    ax.set_xlabel(r'$z$', ha='right', x=1)
    ax.set_ylabel(r'$\lambda(e^z - z^z)$', ha='right', y=1)
    ax.legend(loc='best')
    ax.minorticks_on()
    plt.show()
```

This example contains a bit of customization, and is essentially ready for publication/as a standalone figure

```python
with pretty.Classic():
    fig = plt.figure(figsize = (12,8))    
    ax = fig.add_subplot(1, 1, 1)
    
    ax.plot(x2,y2, label = 'a')
    ax.plot(x2, [z + .1 for z in y2], label = 'b')
    ax.plot(x2, [z + .2 for z in y2], label = 'c')
    ax.plot(x2, [z + .3 for z in y2], label = 'd')        
    ax.set_xlabel(r'$x$', ha='right', x=1)
    ax.set_ylabel(r'$tan(sin(...tan(sin(\gamma x))...))$', ha='right', y=1)
    leg = ax.legend(loc='best', title = r"A bunch" + "\n of lines")
    leg._legend_box.align = "left"
    
    ax.grid(which='minor', alpha=0.4, axis='y', linestyle='dotted')
    ax.grid(which='major', alpha=0.8, linestyle='dotted')
    ax.annotate(r'2019', xy=(1, 1.015), xycoords='axes fraction', ha='right',
                annotation_clip=False, fontsize = 18)
    ax.annotate('$\mathbf{EX.}$', xy=(0, 1.015), xycoords='axes fraction', ha='left',
            fontsize=18, annotation_clip=False)
    ax.annotate('$Plotting \ stylesheet$', xy=(0.055, 1.015), xycoords='axes fraction', ha='left',
            fontsize=18, annotation_clip=False)
    plt.show()
```
